```python
import numpy as np
import pandas as pd
from fbprophet import Prophet
```

```python
df = pd.read_csv("/content/Inflation Rate  (montly).csv")
df.head(2)
```

```python
df['year'] = pd.to_datetime(df['year'])
```

```python
df = df[['year', 'inflation_rate']].reset_index(drop=True)
df.rename(columns={'year':'ds', 'inflation_rate':'y'}, inplace=True)
df.head(2)
```

```python
df
```

```python
train=df.head(345)
test=df.tail(148)
train
```

```python
m = Prophet()
m.fit(train)
```

```python
future = m.make_future_dataframe(periods=148)
future.tail(2)
```

```python
forecast = m.predict(future)
```

```python
forecast.head(2)
```

```python
abc=forecast['yhat']
```

```python
result=abc.tail(148)
result
```

```python
def RMSE(org, pred):
    rmse=np.sqrt(np.mean((np.array(org)-np.array(pred))**2))
    return rmse
```

```python
rmse_prophet_model = RMSE(test['y'], result)
rmse_prophet_model
```
